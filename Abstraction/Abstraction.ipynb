```python
# Example 1: Abstract class for Payment Systems
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def process(self, amount): pass

class CreditCard(Payment):
    def process(self, amount):
        print(f"Processing {amount} via Credit Card")

class PayPal(Payment):
    def process(self, amount):
        print(f"Processing {amount} via PayPal")

for method in [CreditCard(), PayPal()]:
    method.process(500)
```

```python
# Example 2: Abstract class for Notifications
class Notification(ABC):
    @abstractmethod
    def send(self, message): pass

class Email(Notification):
    def send(self, message): print(f"Email sent: {message}")

class SMS(Notification):
    def send(self, message): print(f"SMS sent: {message}")

Email().send("Hello via Email")
SMS().send("Hello via SMS")
```

```python
# Example 3: Abstract class for File Readers
class FileReader(ABC):
    @abstractmethod
    def read(self): pass

class CSVReader(FileReader):
    def read(self): print("Reading CSV file...")

class JSONReader(FileReader):
    def read(self): print("Reading JSON file...")

CSVReader().read()
JSONReader().read()
```

```python
# Example 4: Abstract class for Vehicles
class Vehicle(ABC):
    @abstractmethod
    def start(self): pass

class Car(Vehicle):
    def start(self): print("Car started")

class Bike(Vehicle):
    def start(self): print("Bike started")

Car().start(); Bike().start()
```

```python
# Example 5: Abstract class for Authentication
class Auth(ABC):
    @abstractmethod
    def login(self, user, pwd): pass

class GoogleAuth(Auth):
    def login(self, user, pwd): print(f"{user} logged in via Google")

class LocalAuth(Auth):
    def login(self, user, pwd): print(f"{user} logged in locally")

GoogleAuth().login("Harish", "1234")
LocalAuth().login("Harish", "abcd")
```
