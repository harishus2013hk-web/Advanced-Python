```python
class Bank:
    def __init__(self):
        self.__balance=0
    def deposit(self,amount):
        if amount>0:
            self.__balance+=amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid amount")
    def withdraw(self,amount):
        if amount<=0:
            print("Invalid amount")
        elif amount>self.__balance:
            print("Insufficient balance")
        else:
            self.__balance-=amount
            print(f"Withdrawn amount: {amount}")
    def getbalance(self):
        return self.__balance

account=Bank()
account.deposit(1000)
account.withdraw(400)
print("Current Balance:",account.getbalance())
```

```python
#Getter and Setter Methods
class Employee:
   def __init__(self):
       self.__salary = 50000 # Private attribute
   def get_salary(self): # Getter
       return self.__salary
   def set_salary(self, amount): # Setter
       if amount > 0:
           self.__salary = amount
       else:
           print("Invalid salary amount!")
emp = Employee()
print(emp.get_salary()) # Outputs: 50000
emp.set_salary(60000)
print(emp.get_salary()) # Outputs: 60000
```

```python
class Car:
    def __init__(self, brand, fuel=100):
        self.brand = brand
        self.__speed = 0       # private attribute
        self.__fuel = fuel     # private attribute

    # Public method to accelerate
    def accelerate(self):
        if self.__fuel > 0:
            self.__speed += 10
            self.__fuel -= 5
            print(f"{self.brand} accelerated. Speed: {self.__speed} km/h, Fuel: {self.__fuel}%")
        else:
            print("Out of fuel! Cannot accelerate.")

    # Public method to brake
    def brake(self):
        if self.__speed > 0:
            self.__speed -= 10
            print(f"{self.brand} slowed down. Speed: {self.__speed} km/h")
        else:
            print("Car is already stopped.")

    # Getter for speed
    def get_speed(self):
        return self.__speed

    # Getter for fuel
    def get_fuel(self):
        return self.__fuel

    # Setter for refueling
    def refuel(self, amount):
        if amount > 0:
            self.__fuel += amount
            if self.__fuel > 100:
                self.__fuel = 100
            print(f"{self.brand} refueled. Fuel: {self.__fuel}%")
        else:
            print("Invalid fuel amount.")


# Usage
my_car = Car("Tesla")
my_car.accelerate()
my_car.accelerate()
my_car.brake()
print("Current Speed:", my_car.get_speed())
print("Current Fuel:", my_car.get_fuel())
my_car.refuel(30)
```

```python
class Bird:
    def __init__(self, name, color):
        # Private attributes (encapsulated)
        self.__name = name
        self.__color = color
    
    # Getter methods
    def get_name(self):
        return self.__name
    
    def get_color(self):
        return self.__color
    
    # Setter methods
    def set_color(self, new_color):
        if new_color.lower() in ["red", "blue", "green", "yellow"]:
            self.__color = new_color
        else:
            print("Invalid color! Choose from red, blue, green, yellow.")
    
    # Behavior method
    def fly(self):
        print(f"{self.__name} is flying with {self.__color} feathers!")

# Example usage
sparrow = Bird("Sparrow", "brown")

print(sparrow.get_name())   # Accessing name safely
print(sparrow.get_color())  # Accessing color safely

sparrow.set_color("blue")   # Changing color safely
sparrow.fly()

sparrow.set_color("black") # Invalid color, restricted by encapsulation
```
