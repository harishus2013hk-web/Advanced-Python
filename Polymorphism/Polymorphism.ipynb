```python
class Bird:
    def sound(self):
        print("Some generic bird sound")

class Sparrow(Bird):
    def sound(self):
        print("Chirp chirp")

class Parrot(Bird):
    def sound(self):
        print("Squawk")

for bird in [Sparrow(), Parrot()]:
    bird.sound()

#Different bird classes override the same sound() method.
```

```python
# Polymorphism with different classes Any object with a speak() method works, regardless of its class.
class Dog:
    def speak(self):
        print("Woof!")

class Cat:
    def speak(self):
        print("Meow!")

def animal_sound(animal):
    animal.speak()

animal_sound(Dog())
animal_sound(Cat())
```

```python
# Polymorphism with inheritance and abstract classes. Different shapes implement the same abstract method area().
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14 * self.r * self.r

class Square(Shape):
    def __init__(self, s): self.s = s
    def area(self): return self.s * self.s

for shape in [Circle(5), Square(4)]:
    print(shape.area())
```

```python
# Polymorphism with default arguments. The add() method can take different numbers of arguments.
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(5))        # 5
print(calc.add(5, 10))    # 15
print(calc.add(5, 10, 20))# 35
```

```python
# Polymorphism with method overriding. Subclasses provide their own implementation of the move() method.
class Vehicle:
    def move(self):
        print("Vehicle is moving")

class Car(Vehicle):
    def move(self):
        print("Car drives on road")

class Boat(Vehicle):
    def move(self):
        print("Boat sails on water")

for v in [Car(), Boat()]:
    v.move()
```
