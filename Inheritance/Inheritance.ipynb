```python
#single inheritance example
class Bird:
    def fly(self):
        print("Bird can fly")

class Sparrow(Bird):  # inherits from Bird
    def chirp(self):
        print("Sparrow chirps")

s = Sparrow()
s.fly()   # inherited
s.chirp() # own method
```

```python
#multilevel inheritance example
class Animal:
    def eat(self):
        print("Animal eats")

class Bird(Animal):
    def fly(self):
        print("Bird flies")

class Parrot(Bird):  # inherits from Bird, which inherits from Animal
    def talk(self):
        print("Parrot talks")

p = Parrot()
p.eat()   # from Animal
p.fly()   # from Bird
p.talk()  # from Parrot
```

```python
#hierarchical inheritance example
class Bird:
    def fly(self):
        print("Bird flies")

class Sparrow(Bird):
    def chirp(self):
        print("Sparrow chirps")

class Eagle(Bird):
    def hunt(self):
        print("Eagle hunts")

s = Sparrow()
e = Eagle()
s.fly(); s.chirp()
e.fly(); e.hunt()
```

```python
#multiple inheritance example
class Flyer:
    def fly(self):
        print("Can fly")

class Swimmer:
    def swim(self):
        print("Can swim")

class Duck(Flyer, Swimmer):  # inherits from both
    def quack(self):
        print("Duck quacks")

d = Duck()
d.fly()
d.swim()
d.quack()
```

```python
#hybrid inheritance example
class Animal:
    def eat(self):
        print("Animal eats")

class Bird(Animal):
    def fly(self):
        print("Bird flies")

class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

class Bat(Mammal, Bird):  # hybrid inheritance
    def sound(self):
        print("Bat screeches")

b = Bat()
b.eat()   # from Animal
b.walk()  # from Mammal
b.fly()   # from Bird
b.sound() # from Bat
```
